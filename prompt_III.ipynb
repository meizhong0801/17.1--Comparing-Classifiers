### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features, prepare the features and target column for modeling with appropriate encoding and transformations.

````python
df.head()
````

````python
# Convert target column to numrical
df["y"] = df["y"].map({'no':0, 'yes':1})
df.head()
````

````python
# Categorical columns
string_col = df.select_dtypes(include="object").columns
string_col
````

````python
## Createing one hot encoded features for working with non tree based algorithms 
df_nontree=pd.get_dummies(df,columns=string_col,drop_first=False)
df_nontree.head()
````

````python
## Create label encoder features for working with tree based algorithms
from sklearn.preprocessing import LabelEncoder

df_tree = df.apply(LabelEncoder().fit_transform)
df_tree.head()
````

### Problem 6: Train/Test Split

With your data prepared, split it into a train and test set.

````python
X = df_nontree[[col for col in df_nontree.columns if col != "y"]]
y = df_nontree["y"]
````

````python
X.head()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y)
````

````python
X_train.shape
X_test.shape
````

````python
X_tree = df_tree[[col for col in df_tree.columns if col != "y"]]
y_tree = df_tree["y"]

X_tree_train, X_tree_test, y_tree_train, y_tree_test = train_test_split(X_tree, y_tree)
````

### Problem 7: A Baseline Model

Before we build our first model, we want to establish a baseline.  What is the baseline performance that our classifier should aim to beat?

````python
y = df["y"]
baseline_accuracy = (y == y.mode()[0]).mean()
print(f"Baseline accuracy: {baseline_accuracy:.2f}")
````

### Problem 8: A Simple Model

Use Logistic Regression to build a basic model on your data.  

````python
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression()
clf.fit(X_train,y_train)

````

### Problem 9: Score the Model

What is the accuracy of your model?

````python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score

y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred))
auc=roc_auc_score(y_test,y_pred)
print(f"AUC Score: {auc:.4f}")
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
````
